```python
from svgpathtools import svg2paths, Path, Line, CubicBezier
paths, attributes = svg2paths('/home/xertrov/win/Pictures/Openplanet/hello-openplanet.svg')
print(f"len: {len(paths)}")
```

```python
def path_points_all(el: Line | CubicBezier):
    return [el.start, (el.start + el.end) / 2, el.end]

def path_points_tail(el: Line | CubicBezier):
    return [(el.start + el.end) / 2, el.end]

def path_to_points(path: Path):
    points = []
    for el in path:
        points.extend(path_points_tail(el))
    return points

import json
```

```python
point_lists = [path_to_points(path) for path in paths]
# dir(point_lists[0][0])
point_lists = [[(p.real, p.imag) for p in path_to_points(path)] for path in paths]
json_str = json.dumps(point_lists)
with open('points.json', 'w') as f:
    f.write(json_str)

print(f"wrote: {json_str}")
```
